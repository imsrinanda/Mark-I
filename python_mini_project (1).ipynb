```python
# It is a user defined function to print the attendance of a specific student.
def check(attendance,roll_number):


  option1 = input("Choose an option:\n1. Specific date\n2. Range of dates\n3. Every day in the month\n")
  option1 = int(option1)
  roll_number = int(roll_number)


  if option1 == 1:
    date = input("Enter the date in DD-MM-YYYY format: ")
    student_attendance = attendance[(attendance["Roll Number"] == roll_number) & (attendance["Date"] == date)]
    print(student_attendance)
    return (student_attendance)
  elif option1 == 2:
    start_date = input("Enter the start date in DD-MM-YYYY format: ")
    end_date = input("Enter the end sate in DD-MM-YYYY format: ")
    student_attendance = attendance[(attendance["Roll Number"] == roll_number) & (attendance["Date"].between(start_date, end_date))]
    print(student_attendance)
    return student_attendance
  elif option1 == 3:
    student_attendance = attendance[(attendance["Roll Number"] == roll_number)]
    print(student_attendance)
    return student_attendance
  else:
    print("Invalid option.")
```

```python
# It is a user defined function to check the attendance percentage of a particular student.
def check_attendance_percentage(roll_number):


  roll_number = int(roll_number)
  stud_att = attendance[(attendance["Roll Number"] == roll_number)]
  total_classes = stud_att.shape[0] * 5
  attended_classes = stud_att.iloc[:, 3:].sum().sum()
  leave_classes = total_classes - attended_classes
  attendance_percentage = (attended_classes / total_classes) * 100
  rem_class = ((attendance_percentage/100) - (85/100)) * total_classes

  print(f"\nAttendance Details for roll no. {roll_number}:")
  print(f"\nTotal Classes: {total_classes}")
  print(f"Attended Classes: {attended_classes}")
  print(f"Leave Classes: {leave_classes}")
  print(f"Attendance Percentage: {round(attendance_percentage,2)} %")
  return attendance_percentage,total_classes
```

```python
# Here we are defining a function to check if a student to eligible to write the final examination.
def eligibility(roll_number):

  roll_number = int(roll_number)
  stud_att = attendance[(attendance["Roll Number"] == roll_number)]
  total_classes = stud_att.shape[0] * 5
  attended_classes = stud_att.iloc[:, 3:].sum().sum()
  leave_classes = total_classes - attended_classes
  attendance_percentage = (attended_classes / total_classes) * 100
  rem_class = ((attendance_percentage/100) - (85/100)) * total_classes
  if attendance_percentage >= 85:
    print("This student is eligible to write the final examination!")
  else:
    print("This student cannot write the final examination as they have attendance shortage.")
```

```python
# to find the number of classes that the student can miss

# Below is a user defined function for the same.
def missable_class(roll_number):
  import math


  roll_number = int(roll_number)
  stud_att = attendance[(attendance["Roll Number"] == roll_number)]
  total_classes = stud_att.shape[0] * 5
  attended_classes = stud_att.iloc[:, 3:].sum().sum()
  leave_classes = total_classes - attended_classes
  attendance_percentage = (attended_classes / total_classes) * 100
  rem_class = ((attendance_percentage/100) - (85/100)) * total_classes
  print(f"The Student has attended {attended_classes} out of {total_classes}")
  if rem_class > 0:
    print(f"The student can miss {math.floor(rem_class)} classes.")
  else:
    print(f"The student cannot miss anymore classes.")
```

```python
students = {}

def mark_attendance(name, roll_no, subject, date, is_present):

  if name in students:
        students[name].append((roll_no, subject, date, is_present))
  else:
        students[name] = [(roll_no, subject, date, is_present)]
  # Print the attendance records
  print("Attendance Records:")
  for name, attendance_list in students.items():
      print(f"Student: {name}")
      for roll_no, subject, date, is_present in attendance_list:
          status = "Marked Present" if is_present == 1 else "Marked0Absent"
          print(f"Roll No: {roll_no}, Subject: {subject}, Date: {date}, Status: {status}")
```

```python
op1 = input("No. of time you want to run the program?")
op1 = int(op1)

for i in range(op1):

    op2 = input('''
    Welcome to Attendance Tracking System!

    What do you want to do?
    1. View the attendance details of a specific student.
    2. Check attendance percentage.
    3. Check if a particular student is eligibe to write the exams.
    4. Mark the attendance.
    5. Check the classes the student can miss.
    6. Exit.

    ''')
    op2 = int(op2)
    if op2 == 1:
      roll_number = input("Enter the Roll no. of the student: ")
      check(attendance,roll_number)
    elif op2 == 2:
      roll_number = input("Enter the Roll no. of the student you want to check : ")
      check_attendance_percentage(roll_number)
    elif op2 == 3:
      roll_number = input("Enter the Roll no. of the student you want to check : ")
      eligibility(roll_number)
    elif op2 == 4:
      name = input("Enter the student name: ")
      roll_no = input("Enter the roll no: ")
      subject = input("Enter the subject: ")
      date = input("Enter the date in DD-MM-YYYY format: ")
      is_present = input("To mark the student present, enter 1. To mark the student absent, enter 0: ")
      is_present = int(is_present)
      mark_attendance(name, roll_no, subject, date, is_present)
    elif op2 == 5:
      roll_number = input("Enter the Roll no. of the student you want to check : ")
      missable_class(roll_number)
    elif op2 == 6:
      break
    else:
      print("Invalid input.")
```
